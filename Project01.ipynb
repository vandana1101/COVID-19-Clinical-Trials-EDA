```python
# Function to visualize missing data
def visualize_data(data, caption='', ylabel='Percentage of Missing Data'):
    # Set figure size
    sns.set(rc={'figure.figsize': (15, 8.27)})

    # Make x-axis labels vertical
    plt.xticks(rotation=90)

    # Plot only the first 40 columns (if available)
    top_n = min(40, len(data))  # Ensure we don't exceed available data
    
    # Create bar plot
    fig = sns.barplot(x=list(data.keys())[:top_n], y=list(data.values)[:top_n])
    
    # Set title and labels
    fig.set_title(caption)
    plt.ylabel(ylabel)

    # Show plot
    plt.show()
```

```python
# Visualising the percentage of missing values in each column
visualize_data(missing_data , 'Percentage of missing data')
```

As shown the percentae of missing data in Results First Posted is 99.3% and

```python
# Drop Study Documents and Results First Posted
data.drop(['Results First Posted', 'Study Documents'], axis=1, inplace=True)
```

```python
data.isnull().sum()  #number of null rows in each column after dropping 'Results First Posted' and 'Study Documents'
```

```python
# We can extract a new feature form The Location which is the country where the study hold
from pycountry import countries

# Create a set of valid country names
country_list = set([c.name for c in countries])

# Function to extract country if it's in the country list
def get_country(location):
    for word in str(location).split(","):
        word = word.strip()
        if word in country_list:
            return word
    return "Unknown"

# Apply function
data.loc[:, "Country"] = data["Locations"].apply(get_country)
```

```python
data.Country.value_counts()
```

Now We need to clasify the missing data to one of these categories
1) Missing Completely At Random (MCAR)
2) Missing At Random (MAR)
3) Not Missing At Random (NMAR)

```python
# Lets's start with Acronym
print(f"Number of unique values is {data.Acronym.nunique()} \n")
data.Acronym.value_counts()
```

```python
# Find the realtion between null values in Acronym and Countries
(data.Acronym.isnull().groupby(data.Country).mean().sort_values(ascending = False) * 100)
```

● After inspecting the relation between the missing values in Acronym and
Country we can conclude that there is a sort of relation between these two
features, so we can say that Data is Missing At Random (MAR).
● So we can Impute by Missing Category.

```python
# impute by a missing Indicator
data.Acronym = data.Acronym.fillna("Missing Acronym")
```

```python
# Detecting (Percentage) Missing Data
data.isnull().mean() * 100
```

We can do the same for other categorical features such as Interventions , Phases ,
Locations and other categorical features

```python
categorical_features = data.select_dtypes(include =object).columns
features =categorical_features[data[categorical_features].isnull().mean() >0]
for feature in features:
    data[feature] = data[feature].fillna(f"Missing {feature}")
```

```python
# Detecting (Percentage) Missing Data
data.isnull().mean() * 100
```

```python
# Convert 'Enrollment' column to numeric
data.loc[:, "Enrollment"] = pd.to_numeric(data["Enrollment"], errors="coerce")

# Convert 'Start Date' to datetime
data.loc[:, "Start Date"] = pd.to_datetime(data["Start Date"], errors="coerce")

# Check updated data types
print(data.dtypes)
```

```python
#Now the Time to handle The missing data for the Enrollment
# Check the skewness
data.Enrollment.skew()  #Skewness is 34 #isn't normally distributed
```

```python
# Plotting the distribution of the enrollment
data.Enrollment.plot(kind = 'kde')
```

```python
# we will impute enrollement by its median
# Some Statstical Valuse for the Enrollment Column
min_Value = data.Enrollment.min()
max_Value = data.Enrollment.max()
mean_Value = data.Enrollment.mean()
median_Value = data.Enrollment.median()
std_Value = data.Enrollment.std()
print(f"the min value is {min_Value} \n \
The max value is {max_Value} \n \
The mean is {mean_Value} \n \
The Median is {median_Value} \n \
Standard Devation is {std_Value}")
```

```python
# Using Median to impute Missing Values
data.loc[:,"Enrollment"] = data["Enrollment"].fillna(median_Value)
```

```python
# Detecting (Percentage) Missing Data
data.isnull().mean() * 100
```
